# Import librairies

```python
import requests
import pandas as pd 
import json
```

```python
url_pollution_nantes_2025 = 'https://data.nantesmetropole.fr//api/explore/v2.1/catalog/datasets/323266205_indice-atmo-nantes-metropole/records?limit=100&timezone=Europe%2FParis&refine=lib_zone%3A%22Nantes%22&refine=date_ech%3A%222025%22'
```

```python
response = requests.get(url_pollution_nantes_2025)
if response.status_code == 200 :
    data_json = response.json()
    if data_json :
        data = data_json['results']
```

# Processing

```python
data_dict_list = []

for item in data:
    data_dict = {
        "gml_id": item["gml_id"],
        "date_ech": item["date_ech"],
        "code_qual": item["code_qual"],
        "lib_qual": item["lib_qual"],
        "coul_qual": item["coul_qual"],
        "date_dif": item["date_dif"],
        "source": item["source"],
        "type_zone": item["type_zone"],
        "code_zone": item["code_zone"],
        "lib_zone": item["lib_zone"],
        "code_no2": item["code_no2"],
        "code_so2": item["code_so2"],
        "code_o3": item["code_o3"],
        "code_pm10": item["code_pm10"],
        "code_pm25": item["code_pm25"],
        "x_wgs84": item["x_wgs84"],
        "y_wgs84": item["y_wgs84"],
        "x_reg": item["x_reg"],
        "y_reg": item["y_reg"],
        "epsg_reg": item["epsg_reg"],
        "etat_indice": item["etat_indice"],
        "geom_type": item["geom"]["type"],  # Type de géométrie
        "geom_coordinates": item["geom"]["geometry"]["coordinates"],  # Coordonnées géométriques
        "geo_point_2d_lon": item["geo_point_2d"]["lon"],  # Longitude
        "geo_point_2d_lat": item["geo_point_2d"]["lat"]   # Latitude
    }
    data_dict_list.append(data_dict)

# Conversion en DataFrame
df = pd.DataFrame(data_dict_list)
```
